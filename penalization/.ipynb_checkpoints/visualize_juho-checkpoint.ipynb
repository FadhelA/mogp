```python
from sampling_utils import lam_dist, GBFRYInit

p = 1000
ckpts = torch.load(f'results/sghmc/horseshoe_ffnn_{p}_v2.net')
horseshoe_rvs = lam_dist(p, 'horseshoe')
avg_weight = 0
for ckpt in ckpts:
    #avg_weight += horseshoe_rvs.map_to_domain(ckpt['output_layer.transformed_variances']).cpu().data.numpy()
    #avg_weight += ckpt['output_layer.weight'].cpu().data.numpy()

    lam = horseshoe_rvs.map_to_domain(ckpt['output_layer.transformed_variances']).cpu().data.numpy()
    V = ckpt['output_layer.weight'].cpu().data.numpy()
    avg_weight += (lam**0.5)*V

avg_weight /= len(ckpts)
print('Output layer weight dist for horseshoe')
plt.hist(avg_weight.reshape(-1), edgecolor='k', facecolor='b', alpha=0.4, bins=20);
```

```python
models = ['normal', 'horseshoe', 'gbfry_light', 'gbfry_heavy']
num_uniques = {p: {model:[] for model in models} for p in [500, 1000, 2000]}
num_elems = [7, 14, 21, 30]


for i, p in enumerate([500, 1000, 2000]):
    for model in models:
        for num_elem in num_elems:
            num_uniques[p][model].append(len(np.unique(idxs[model][i][:,:num_elem])))       
        
for p in [500, 1000, 2000]:
    plt.figure(figsize=(6,4))
    plt.title(rf'$p={p}$')
    plt.xlabel('number of images / neuron')
    plt.ylabel('number of unique images')
    markers = ['o', 'x', '+', 'v']
    for model, marker in zip(models, markers):
        plt.plot(num_elems, num_uniques[p][model],            
            marker=marker,
            label=model)
    plt.legend()
```

```python
import torch.nn as nn
from fast_nist import FastFMNIST

num_neurons = 30
testdata = FastFMNIST("../data", train=False, download=True, device='cuda')
testset = torch.utils.data.DataLoader(testdata, batch_size=1000)

def test_model(X, y, X_test, Y_test, num_train):
    mlp = nn.Sequential(
            nn.Linear(num_neurons, 100), nn.ReLU(),
            nn.Linear(100, num_neurons)).cuda()
    loss_fn = nn.CrossEntropyLoss(reduction='mean')
    X = torch.Tensor(X).cuda()[:num_train]
    y = torch.LongTensor(y.reshape(-1)).cuda()[:num_train]
 
    opt = torch.optim.Adam(params=mlp.parameters(), lr=1e-4)
    num_iter = 1000
    for i in range(1, num_iter+1):
        mlp.zero_grad()
        logits = mlp(X)
        loss = loss_fn(logits, y)
        loss.backward()
        opt.step()
        # if i % 100 == 0:
        #     print(f'{model} iter {i} loss {loss.item():.4f}')

    with torch.no_grad():
        train_acc = np.round((mlp(X).argmax(-1)==y).float().mean().item(), 4)

    X_test = torch.Tensor(X_test).cuda()
    Y_test = Y_test.cuda()
    with torch.no_grad():
        test_acc = np.round((mlp(X_test).argmax(-1)==Y_test).float().mean().item(), 4)

    # num_correct = 0
    # with torch.no_grad():
    #     test_acc = np.round((mlp(X_test).argmax(-1)))
    #     for x, y in testset:
    #         logits = mlp(x.cuda().reshape(-1, 784))
    #         num_correct += (logits.argmax(-1)==y).sum()
    # test_acc = np.round((num_correct/len(testdata)).item(), 4)

    return train_acc, test_acc    

models = ['normal', 'horseshoe', 'gbfry_light', 'gbfry_heavy']
num_trains = [200, 400, 600, 800]
accs = {p: {model:[] for model in models} for p in [500, 1000, 2000]}
for i, p in enumerate([500, 1000, 2000]):    
    for model in models:                       
        X = features[model][i].reshape(-1, num_neurons)
        Y = labels[model][i].reshape(-1)
        X_test = test_features[model][i]
        for num_train in num_trains:           
            acc = test_model(X, Y, X_test, Y_te, num_train)
            accs[p][model].append(acc)            
```

```python
import matplotlib.pyplot as plt
for p in [500, 1000, 2000]:
    plt.figure(figsize=(6,4))
    plt.title(rf'$p={p}$')
    plt.xlabel('number of training data')
    plt.ylabel('test accuracy')
    markers = ['o', 'x', '+', 'v']
    for model, marker in zip(models, markers):
        plt.plot(num_trains, 
            [test_acc for (train_acc, test_acc) in accs[p][model]], 
            marker=marker,
            label=model)
    plt.legend()
```
